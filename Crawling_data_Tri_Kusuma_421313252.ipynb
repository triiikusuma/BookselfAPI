MENGGUNAKAN BEAUTIFUL SOUP

Beautiful Soup adalah sebuah perpustakaan yang memudahkan dalam mengikis informasi dari halaman web. Perpustakaan ini berada di atas parser HTML atau XML, menyediakan idiom Pythonic untuk mengulangi, mencari, dan memodifikasi pohon parse.

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/Big_data'
```

```python
#package meminta ke url, kemudian mengirim mendapatkan respon dari web
r = requests.get(url)
r = r.text #mengembalikan HTML ke halaman web sebagai string
```

```python
#scrape informasi dari url yang telah diberikan
soup = BeautifulSoup(r)
```

```python
print(soup)
```

```python
#membuat output agar menjadi lebih baik
pretty = soup.prettify
```

```python
print(pretty)
```

```python
#Mendapatkan teks tertentu(spesifik)

#Mengambil judul di sebuah halaman/website
title = soup.title
print(title)
```

```python
#Mendapatkan teks dari artikel
text = soup.get_text()
print(text)
```

```python
datas = []
count_page = 0
for page in range(1,11):
    count_page+=1
    print('scraping page:',count_page)
    url = f'https://www.freelancer.co.id/jobs/{page}/?search_keyword=data%20analyst&search_location=Pekerjaan%20Online'
    req = requests.get(url)
    soup = BeautifulSoup(req.text,'html.parser')
    items = soup.find_all("div", class_="JobSearchCard-primary")
    for it in items:
        name = it.find("a", class_="JobSearchCard-primary-heading-link").text.strip()
        try:
          price = it.find("div",class_="JobSearchCard-primary-price").text.strip()
        except:
          continue
        datas.append([name,price])
        # print(seller_name)
```
